```
from pyspark.sql.types import StructType, StructField, IntegerType, StringType

# Surowe Dane
raw_data = spark.read.text("/databricks-datasets/adult/adult.data")
raw_data.show(5, truncate=False)

# Schema
adult_schema = StructType([
    StructField("age", IntegerType(), True),
    StructField("workclass", StringType(), True),
    StructField("fnlwgt", IntegerType(), True),
    StructField("education", StringType(), True),
    StructField("education_num", IntegerType(), True),
    StructField("marital_status", StringType(), True),
    StructField("occupation", StringType(), True),
    StructField("relationship", StringType(), True),
    StructField("race", StringType(), True),
    StructField("sex", StringType(), True),
    StructField("capital_gain", IntegerType(), True),
    StructField("capital_loss", IntegerType(), True),
    StructField("hours_per_week", IntegerType(), True),
    StructField("native_country", StringType(), True),
    StructField("income", StringType(), True)
])


# Read data
adult_df = spark.read.csv(
    path="/databricks-datasets/adult/adult.data",
    schema=adult_schema,
    header=False,
    sep=", ",  
    inferSchema=False,
    ignoreLeadingWhiteSpace=True,
    ignoreTrailingWhiteSpace=True
)


display(adult_df.head(10))
adult_df.printSchema()
```

# Zadanie 3

```
from pyspark.sql.types import *
df1 = spark.read.format("csv").schema(adult_schema).options(
    sep=",", 
    header=True, 
    mode="PERMISSIVE",
    ignoreLeadingWhiteSpace=True,
    ignoreTrailingWhiteSpace=True
    ).load("/databricks-datasets/adult/adult.data")
df2 = spark.read.format("csv").schema(adult_schema).options(
    sep=",", 
    header=True, 
    mode="DROPMALFORMED",
    ignoreLeadingWhiteSpace=True,
    ignoreTrailingWhiteSpace=True
    ).load("/databricks-datasets/adult/adult.data")

df3 = spark.read.format("csv").schema(adult_schema).options(
    sep=",", 
    header=True, 
    mode="FAILFAST",
    ignoreLeadingWhiteSpace=True,
    ignoreTrailingWhiteSpace=True
    ).load("/databricks-datasets/adult/adult.data")

# podgladam jak wygladaja dane
display(df1.head(20))
display(df2.head(20))
display(df3.head(20))
```

## Psucie danych

```
from pyspark.sql.functions import *
from pyspark.sql.types import StringType

df1 = df1.withColumn(
    "age", 
    when(df1.age > 50, "Older").otherwise(df1.age)
)
df1 = df1.withColumn(
    "capital_gain", 
    when(df1.capital_gain == 0, None).otherwise(df1.capital_gain)
)
df1 = df1.withColumn(
    "education_num", 
    when(df1.education_num < 10, "uneducated").otherwise(df1.capital_loss)
)

df1.write.mode("overwrite").csv("dbfs:/tmp/df1_malformed.csv", sep=",", header=True)
```
